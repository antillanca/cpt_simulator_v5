```
# 2. Definir Red Estándar CPT
class TabularNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.net(x)

model = TabularNet()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.BCELoss()
```

```
# 3. Entrenar
X = torch.tensor(df[['f0', 'f1']].values, dtype=torch.float32)
y = torch.tensor(df['success'].values, dtype=torch.float32).view(-1, 1)

for epoch in range(200):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()
    if epoch % 50 == 0: print(f'Epoch {epoch}, Loss: {loss.item():.4f}')

torch.save(model.state_dict(), 'logic_tabular_filter.pt')
print('✅ Modelo Lógico Guardado')
```
